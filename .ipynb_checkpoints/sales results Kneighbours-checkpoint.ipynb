```python
#import necessary modules
import pandas as pd, matplotlib.pyplot as plt, numpy as np, seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import GaussianNB
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import confusion_matrix, accuracy_score, \
roc_auc_score, plot_confusion_matrix, classification_report

from yellowbrick.classifier import ClassificationReport, ConfusionMatrix

# Read in the data with `read_csv()`
sales_data = pd.read_csv("WA_Fn-UseC_-Sales-Win-Loss.csv")
# Print shape of sales_data
print(sales_data.shape)
sales_data.head()
sales_data.columns
```

```python
# show the different label counts (won vs loss) in % too
print(sales_data['Opportunity Result'].value_counts())
print(sales_data['Opportunity Result'].value_counts(normalize=True))

# show first 5 entries opporunity result
print(sales_data['Opportunity Result'].head())

# create the Labelencoder object
le = preprocessing.LabelEncoder()

#storing the categorical targets win/loss for labelling later
origlabels = sales_data['Opportunity Result'].unique()
#convert the categorical target column into numeric
sales_data['Opportunity Result'] = le.fit_transform(sales_data['Opportunity Result'])

# show first 5 entries opporunity result after label encoding
print(sales_data['Opportunity Result'].head())

#display the initial records
sales_data.head()

# select columns other than 'Opportunity Number','Opportunity Result'
feature_cols = [col for col in sales_data.columns \
                if col not in ['Opportunity Number','Opportunity Result']]
# dropping the 'Opportunity Number'and 'Opportunity Result' columns
data = sales_data[feature_cols]
#assigning the Oppurtunity Result column as target
target = sales_data['Opportunity Result']
dataohc=pd.get_dummies(data)
```

```python
#split data set into train and test sets
data_train, data_test, target_train, target_test = train_test_split(dataohc,target, test_size = 0.30, random_state = 10)
# check ratio of label split
print("Target train labels ratio:\n", target_train.value_counts(normalize=True))
print("Target test labels ratio:\n", target_test.value_counts(normalize=True))

sc = preprocessing.StandardScaler()
X_train = sc.fit_transform(data_train)
X_test = sc.transform(data_test)

classifiers = [
    GaussianNB(),
    LinearSVC(random_state=0, max_iter=1000),
    KNeighborsClassifier(n_neighbors=5)]


# Logging for Visual Comparison

class_al =[]
acc1 = []
confusion_matrix_list = []
translabels=le.fit_transform(origlabels)

for clf in classifiers:
    clf.fit(data_train, target_train)
    name = clf.__class__.__name__
    class_al.append(name)
    test_predictions = clf.predict(data_test)
    acc = accuracy_score(target_test, test_predictions)
    print(f"{name} accuracy : {acc}")
    acc1.append(acc)
    matrix = confusion_matrix(target_test, test_predictions,labels=translabels)
    print(matrix)
    confusion_matrix_list.append(matrix)
    
max_acc = max(acc1) 
fig, ax = plt.subplots(figsize=(10,10))    
ax.bar(class_al,acc1)
ax.set_xlabel('Classification Al')
ax.set_ylabel('Accuracy')
ax.set_title('Classifier Accuracy')
for tick in ax.get_xticklabels():
    tick.set_rotation(90)
    
ax.axhline(y = max_acc,color = "red",linestyle="--")
ax.annotate("Baseline Acc", xy=(4.5,max_acc))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 10), sharey=False)

fig.add_subplot(111, frameon=False)
plt.grid(False)
plt.tick_params(labelcolor='none', top='off', bottom='off', left='off', right='off')
plt.xlabel("Ground Truth", size=30)
plt.ylabel("Prediction", size=30)

labels = origlabels
fig.suptitle('Confusion Matrix for diff models', fontsize=30)
for i in range(len(confusion_matrix_list)):
    sns.heatmap(confusion_matrix_list[i],annot=True, cmap="coolwarm", fmt='d', ax=ax[i],cbar=None,annot_kws={"size": 30})
    ax[i].set_title(f"{class_al[i]}", size=20)
    ax[i].set_xticklabels(labels, size=20);
    ax[i].set_yticklabels(labels, size=20);
#     ax[i].set_ylabel('Prediction', size=30);
#     ax[i].set_xlabel('Ground Truth', size=30)
# sns.heatmap(confusion_matrix_list[0],annot=True, cmap="coolwarm", fmt='d', ax=ax[0])
# labels = ['Loss', 'Won']
# ax[0].set_xticklabels(labels);
# ax[0].set_yticklabels(labels);
# ax[0].set_ylabel('Prediction');
# ax[0].set_xlabel('Ground Truth')
```

```python
# After dropping all the highly correlated columns, 
# realised that the labels was skewed towards losses
# and a stratifiedshufflesplit to maintain ratio when splitting might yield better results

from sklearn.model_selection import StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits=1, test_size=.3, random_state=42)

# Creating empty dataframes
X_train, y_train = pd.DataFrame(), pd.DataFrame()
X_test, y_test = pd.DataFrame(), pd.DataFrame()

# Populating the data and target sets
for train_index, test_index in sss.split(dataohc, target):
    X_train=dataohc.iloc[train_index]
    y_train=target.iloc[train_index]   
    X_test=dataohc.iloc[test_index]
    y_test=target.iloc[test_index]


# Check ratio of opportunity result labels afgter stratified shuffle split
print(y_train.value_counts(normalize=True))
print(y_test.value_counts(normalize=True))

#create object of the classifier
neighsss = KNeighborsClassifier(n_neighbors=5)
#Train the algorithm
neighsss.fit(X_train, y_train)
# predict the response
predknnsss = neighsss.predict(X_test)
# evaluate accuracy
print ("KNeighbors accuracy score w SSS: ",accuracy_score(y_test, predknnsss))

fig, axs = plt.subplots(nrows=1, ncols=2, sharey=False, figsize=(12, 6))

sns.heatmap(confusion_matrix_list[2],annot=True, cmap="coolwarm", fmt='d', ax=axs[0],cbar=None)
axs[0].set_title(f"Original {class_al[2]}", size=20)
axs[0].set_xticklabels(origlabels)
axs[0].set_yticklabels(origlabels)

sns.set_context('talk')
cma = confusion_matrix(y_test, predknnsss, labels = translabels)
axs[1].set_title("After SSS")
ax1 = sns.heatmap(cma, annot=True, cmap="coolwarm", fmt='d', ax=axs[1])
ax1.set_xticklabels(origlabels);
ax1.set_yticklabels(origlabels);
ax1.set_ylabel('Prediction');
ax1.set_xlabel('Ground Truth')
```
